```python
import pandas as pd
import numpy as np
from sklearn import datasets

df = pd.read_csv("symtomps.csv")

FEATURE_NAMES = df.columns.tolist()
FEATURE_NAMES.pop(0)
FEATURE_NAMES.pop(0)
df = df.drop(0, axis=0)
X = df.drop(['ID_Treatment','Symptoms'], axis=1).fillna(0)
y = pd.DataFrame(df["ID_Treatment"])
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
```
